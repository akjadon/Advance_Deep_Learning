# Load a saved TensorFlow model

### Problem Statement
#### Load a saved neural network model that classifies [images of clothing, like sneakers and shirts](https://github.com/zalandoresearch/fashion-mnist) and evaluate the loaded model.

### Load the libraries

```python
import keras
import tensorflow as tf
from keras.models import load_model
```

### Load the model

Your saved model can  be loaded by calling the load_model() function and passing the filename. The function returns the model with the same architecture and weights.


```python
model = load_model('model.h5')
```

```python
model.summary()
```

### Load the test data

```python
fashion_mnist = keras.datasets.fashion_mnist
(_, _), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# scale the values to 0.0 to 1.0
test_images = test_images / 255.0
```

```python
# reshape for feeding into the model
test_images = test_images.reshape(test_images.shape[0], 28, 28, 1)
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
print('test_images.shape: {}, of {}'.format(test_images.shape, test_images.dtype))
```

### Evaluate the model
Evaluate the model on the same dataset to confirm the weights and architecture are the same

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('\nTest accuracy: {}'.format(test_acc))
```
